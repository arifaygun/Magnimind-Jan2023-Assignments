### 8- Create a DataFrame whose index is the first 10 letters of the alphabet and contains two more columns with first 10 prime numbers and the first 10 fibonacci numbers.

```python
Prime_number = []

for i in range(100):
  L = []
  for j in range(2,i):
    if i%j == 0:
      L.append(i)
  if len(L) == 0 and i!=0 and i!=1:
    Prime_number.append(i)

Prime_number[:10]
```

```python
fibonacci_numbers = [0,1]

i = 0
while i<8:
  fibonacci_numbers.append(fibonacci_numbers[-1]+fibonacci_numbers[-2])
  i += 1

fibonacci_numbers  
```

```python
import string
string.ascii_uppercase
```

```python
list(string.ascii_uppercase[:10])
```

```python

df = pd.DataFrame({'Prime_number' : Prime_number[:10],
                   'fibonacci_numbers' : fibonacci_numbers},
                  index=list(string.ascii_uppercase[:10]))
df
```
