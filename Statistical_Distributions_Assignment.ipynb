##### a) Drawing the histogram and interpret

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
distribution = np.random.normal(0, 1, 200000)
```

```python
plt.hist(distribution,50 , density=True)
plt.show()
```

```python
distribution.mean()
```

```python
distribution.std()
```

We observe that the data shows symmetric (bell-curved) distribution and most frequent observations gather around mean. Mean = 0, Std = 1.

##### b) Drawing Q-Q plot and interpret

```python
from statsmodels.graphics.gofplots import qqplot
import pylab as py
```

```python
plt.figure(figsize=(10,10))
qqplot(distribution, line ="45")
plt.title("Plot of Q-Q")
py.show()
```

##### The blue points moving too closely to the 45 degree line, which is nice because it tells us that our data follows normal distribution.
