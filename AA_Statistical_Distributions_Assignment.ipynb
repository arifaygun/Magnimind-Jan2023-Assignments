```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.graphics.gofplots import qqplot
```

```python
mean, std = 2,3

phi_X1 = np.random.normal(loc=mean, scale=std, size=100)

fig, ax = plt.subplots(2,2, figsize=(10,8))
bins=20
sns.histplot(phi_X1, kde=True , bins=bins, ax=ax[0,0])

s = pd.Series(phi_X1).value_counts(bins=bins).values.max()
ax[0,0].plot(mean*np.ones(s), range(s), color='r', linestyle='--')
ax[0,0].plot(mean*np.ones(s)+std, range(s), color='k', linestyle='--')
ax[0,0].plot(mean*np.ones(s)-std, range(s), color='k', linestyle='--')
ax[0,0].set_title('Normal Distribution', fontsize = 14)

qqplot(phi_X1, line='45', ax=ax[0,1])

phi_X1 = (phi_X1 - mean)/std
sns.histplot(phi_X1, kde=True, bins=20, ax = ax[1,0])
s = pd.Series(phi_X1).value_counts(bins=bins).values.max()
ax[0,0].plot(mean*np.ones(s), range(s), color='r', linestyle='--')
ax[0,0].plot(mean*np.ones(s)+std, range(s), color='k', linestyle='--')
ax[0,0].plot(mean*np.ones(s)-std, range(s), color='k', linestyle='--')
ax[0,0].set_title('Normal Distribution', fontsize = 14)

qqplot(phi_X1, line='45', ax=ax[1,1])

plt.tight_layout(w_pad = 3)
plt.show()
```

```python
plt.figure(figsize=(20,10))

k=2
mean=2
std=3

for i in range(2,6):
  phi_X1 = np.random.normal(loc=mean, scale=std, size=10**i)
  bins=50
  plt.subplot(2,k,i-1)
  sns.histplot(phi_X1, kde=True, bins=bins)
  plt.xlabel(f'sample+size = {10**i}', fontsize=16)
  plt.xticks(range(round(phi_X1.min()), round(phi_X1.max())))
  s=pd.Series(phi_X1).value_counts(bins=bins).values.max()
  plt.plot(mean*np.ones(s), range(s), color='r', linestyle='--')
  plt.plot(mean*np.ones(s)+std, range(s), color='k', linestyle='--')
  plt.plot(mean*np.ones(s)-std, range(s), color='k', linestyle='--')
  sr=pd.Series(phi_X1)
  plt.title(f'{len(sr[(sr>=mean-std) & (sr<=mean+std)])*100/len(sr)} % of data is between -std = {mean-std} and +std= {mean+std}\n', fontsize=18)

plt.tight_layout(w_pad=2)
plt.show()
```
